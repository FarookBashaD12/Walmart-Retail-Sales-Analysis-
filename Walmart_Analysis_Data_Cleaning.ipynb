```python
# To convert Date columns to datetime
df['Date'] = pd.to_datetime(df['Date'])
ft['Date'] = pd.to_datetime(ft['Date'])
```

```python
#to convert store and dept to integer 
df["Store"] = df["Store"].astype(int)
df["Dept"] = df["Dept"].astype(int)
```

```python
# To Handle Missing Values, we have two option 
#option 1: Fill NA  with 0 for markdown
ft.fillna({'MarkDown1': 0, 'MarkDown2': 0, 'MarkDown3': 0, 'MarkDown4': 0, 'MarkDown5': 0}, inplace=True)
```

```python
# Merge train and features on ['Store', 'Date']
merged_ab = pd.merge(df, ft, on=['Store', 'Date'], how='left')

# Merge the result with stores on 'Store'
final_ab = pd.merge(merged_ab, st, on='Store', how='left')

# Inspect final dataset
print(final_ab.shape)
print(final_ab.columns)
```

```python
final_ab['Month'] = final_ab['Date'].dt.to_period('M')
monthly_sales = final_ab.groupby('Month')['Weekly_Sales'].sum()
print(monthly_sales)
```

```python
store_perf = final_ab.groupby('Store')['Weekly_Sales'].sum().sort_values(ascending=False)
print(store_perf)
```

```python
# Convert 'Date' to datetime if not done yet
df['Date'] = pd.to_datetime(df['Date'])
ft['Date'] = pd.to_datetime(ft['Date'])

# Merge train and features
ab = pd.merge(df, ft, on=['Store', 'Date'], how='left')

# Merge with stores
final_ab = pd.merge(ab, st, on='Store', how='left')

# Check again
print('IsHoliday' in final_ab.columns)  # Should return True

```

```python
# Rename 'IsHoliday_y' to 'IsHoliday' for clarity
ab.rename(columns={'IsHoliday_y': 'IsHoliday'}, inplace=True)

# Now merge with stores
final_ab = pd.merge(ab, st, on='Store', how='left')

# Confirm column exists
print('IsHoliday' in final_ab.columns)  # Should return True
```

```python
holiday_sales = final_ab.groupby('IsHoliday')['Weekly_Sales'].mean()
print(holiday_sales)
```

```python
dept_sales = final_ab.groupby('Dept')['Weekly_Sales'].sum().sort_values(ascending=False)
print(dept_sales)
```

```python
type_sales = final_ab.groupby('Type')['Weekly_Sales'].mean()
size_sales = final_ab.groupby('Size')['Weekly_Sales'].mean()

print(type_sales)
print(size_sales)
```
